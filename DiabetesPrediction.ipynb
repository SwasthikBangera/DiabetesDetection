Import Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score

```

Data import and Analysis

```python
#Loading dataset to Pandas
diabetic_ds = pd.read_csv('/content/diabetes.csv')
```

```python
#Read the data
diabetic_ds.head()
```

```python
print(diabetic_ds.shape)
```

```python
#Statistical data of the data
diabetic_ds.describe()
```

```python
diabetic_ds['Outcome'].value_counts()
```

0 is non-diabetic and 1 shows diabetic

```python
diabetic_ds.groupby('Outcome').mean()
```

```python
# Data seperated from labels
X = diabetic_ds.drop(columns = 'Outcome', axis =1)
Y = diabetic_ds['Outcome']
```

```python
X.head()
```

```python
Y.head()
```

```python
#Standardize the data
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
std_data = scaler.transform(X)
```

```python
print(std_data)
```

```python
X = std_data
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_test.shape, X_train.shape)
```

```python
#Training the model
classifier = svm.SVC(kernel='linear')

classifier.fit(X_train, Y_train)
```

```python
#Evaluating the model
#Accuracy score on training data
X_train_prediction = classifier.predict(X_train)
traindata_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy score of training data: ', traindata_accuracy)
```

```python
#Accuracy score on test data
X_test_prediction = classifier.predict(X_test)
testdata_accuracy = accuracy_score(X_test_prediction, Y_test)

print('Accuracy score of test data: ', testdata_accuracy)
```

```python
#Prediction on new data selected at random

input_data = (1,103,30,38,83,43.3,0.183,33)

#change input data as numpy array
input_data_np_array = np.asarray(input_data)

#reshape array to predict for a single instance
input_data_reshaped = input_data_np_array.reshape(1,-1)

#Standardized inout data
std_data = scaler.transform(input_data_reshaped)

print(std_data)

prediction = classifier.predict(std_data)
print(prediction)


if (prediction [0] == 0):
  print('The person is not diabetic')
else:
  print('The person is diabetic')
```
